# Transformation in ETL

1. Define Data Transformation in ETL and explain why it is important.

- Transformation is the stage in the ETL process where raw, unorganized inconsistent, or incomplete data is systematically converted into
clean, standardized, and analysis-ready formats.It involves applying a set of business rules,
calculations, cleaning techniques, and
restructuring methods to ensure that the data is
meaningful, accurate, and usable for
decision-making.

Data transformation ensures data quality by cleaning, standardizing, and normalizing raw data to remove typing mistakes, incomplete entries, and duplicates; it prepares datasets for analysis by structuring and typing data for visualization; it enables machine learning through encoding text into numbers and scaling values; and it reduces errors in reporting.

3. What is the difference between Normalization and Standardization?

- Distribution Assumption:
Standardization assumes data follows a Gaussian (normal) distribution. Normalization does not assume any specific data distribution.
Outlier Sensitivity:
Standardization is less affected by outliers because it doesn't bound data to a fixed range. Normalization can be heavily impacted by extreme outliers, which skew the scaling range.
Definition & Formula:
Standardization centers data around O with a standard deviation of 1, using the Z-score formula and Normalization scales data, often between 0 and 1.

8. Explain why Z-score Standardization is preferred over Min-Max Scaling when outliers exist.

- Z-score standardization is preferred over Min-Max scaling when outliers exist because Min-Max scaling compresses the "normal" data into an extremely small, almost unusable range, while Z-score maintains the distribution of the majority of data points. If the dataset contains outliers (e.g., extremely large or small values), these outliers determine the maximum or minimum value, severely skewing the scaling for all other data points.In case of Z-score Standardization while the mean and standard deviation are still affected by outliers, the impact is mitigated compared to the direct influence of a minimum/maximum value.
